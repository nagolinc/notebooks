```python
from apples import simulation
```

```python
import matplotlib.pyplot as plt

def plot_simulation_results(results):
    steps = []
    pear_trees = []
    apple_trees = []
    pear_laborers = []
    apple_laborers = []
    pears_picked = []
    apples_picked = []
    populations = []
    profits = []
    
    for result in results:
        steps.append(result['step'])
        pear_trees.append(result['num_pear_trees'])
        apple_trees.append(result['num_apple_trees'])
        pear_laborers.append(result['num_pear_laborers'])
        apple_laborers.append(result['num_apple_laborers'])
        pears_picked.append(result['num_pears_picked_today'])
        apples_picked.append(result['num_apples_picked_today'])
        populations.append(result['population'])
        profits.append(result['profit'])
        
    plt.figure(figsize=(14, 10))

    plt.subplot(321)
    plt.plot(steps, pear_trees, label='Pear Trees')
    plt.plot(steps, apple_trees, label='Apple Trees')
    plt.title('Number of Trees')
    plt.legend()

    plt.subplot(322)
    plt.plot(steps, pear_laborers, label='Pear Laborers')
    plt.plot(steps, apple_laborers, label='Apple Laborers')
    plt.title('Number of Laborers')
    plt.legend()

    plt.subplot(323)
    plt.plot(steps, pears_picked, label='Pears Picked')
    plt.plot(steps, apples_picked, label='Apples Picked')
    plt.title('Fruits Picked')
    plt.legend()

    plt.subplot(324)
    plt.plot(steps, populations, label='Population')
    plt.title('Population')
    plt.legend()

    plt.subplot(325)
    plt.plot(steps, profits, label='Profit')
    plt.title('Profit')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Parameters for the simulation, modify these as necessary
simulation_parameters = {
    'max_apple_trees': 500,
    'max_pear_trees': 500,
    'num_apple_trees': 100,
    'num_pear_trees': 100,
    'pears_per_tree': 4,
    'apples_per_tree': 4,
    'population': 1,
    'p_tree_survival': 0.99,
    'healthy_fruit_amount': 10,
    'population_growth_rate': 1.1,
    'hours_per_day': 1,
    'search_multiplier':40,
    'num_steps':500,
    'epsilon':1,
}

# Run the simulation
results = simulation(**simulation_parameters,verbose=False)

# Plot the results
plot_simulation_results(results)
```
